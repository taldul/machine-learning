First we are going to import some libraries.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_rows', 10)
pd.set_option('display.max_columns', 20)
pd.set_option('display.width', 1000)
pd.options.display.max_columns=300
```

Scikit learn or Sklearn is the primary machine learning library we will use in this course. 

In additionת Sklearn contains datasets and we are going to look at one of these datasets.

```python
import sklearn
from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer()
print(type(data))
```

We will convert the scikit learn dataset to a dataframe.

```python
def sklearn_to_df(sklearn_dataset):
    df = pd.DataFrame(sklearn_dataset.data, columns=sklearn_dataset.feature_names)
    df['target'] = pd.Series(sklearn_dataset.target)
    return df
```

```python
wisc_cancer = sklearn_to_df(data)
```

<h2> Visualizations </h2>

```python
sns.set(rc={'figure.figsize':(6,6)})
sns.countplot(x=wisc_cancer['target'], 
              label='count', 
              palette=['firebrick','seagreen'])
plt.show()
```

```python
sns.set(rc={'figure.figsize':(6,6)})
sns.set_theme(style="whitegrid", palette="pastel")
ax = sns.countplot(x=wisc_cancer['target'], 
              label='count')
ax.set(xlabel='common xlabel', ylabel='common ylabel')
ax.set_title('Title of Graph')
plt.show()
```

```python
data = pd.concat([wisc_cancer['target'], wisc_cancer.iloc[:,1:2]], axis=1)
data2 = pd.melt(data, id_vars="target", var_name="features", value_name='value')

display(data2.head())
```

```python
sns.violinplot(x="features", y="value", hue="target", data=data2, split=True)
```

```python
my_pal = {0: "firebrick", 1: "navy"}

data = pd.concat([wisc_cancer['target'], wisc_cancer.iloc[:,1:3]], axis=1)
data2 = pd.melt(data, id_vars="target", var_name="features")

sns.violinplot(x="features", y="value", hue="target", data=data2, split=True, palette=my_pal)
plt.show()
```

```python
features = ((wisc_cancer.iloc[:,1:3] - wisc_cancer.iloc[:,1:3].mean()) / wisc_cancer.iloc[:,1:3].std())

my_pal = {0: "firebrick", 1: "navy"}

data = pd.concat([wisc_cancer['target'], features], axis=1)
data2 = pd.melt(data, id_vars="target", var_name="features")



sns.violinplot(x="features", y="value", hue="target", data=data2, split=True, palette=my_pal)
plt.show()
```
