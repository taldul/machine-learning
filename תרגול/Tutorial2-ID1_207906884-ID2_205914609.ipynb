<strong> Assignment #2 </strong>

Construct a **histogram** of sepal length, devided by iris specie.

```python
x1=iris.loc[iris.variety=='Virginica','sepal.length']
x2=iris.loc[iris.variety=='Setosa','sepal.length']
x3=iris.loc[iris.variety=='Versicolor','sepal.length']


plt.hist(x1, label='Virginica')
plt.hist(x2, label='Setosa')
plt.hist(x3, label='Versicolor')

plt.gca().set(title='Frequency Histogram of sepal length, devided by iris specie',xlabel='sepal length', ylabel='Frequency')
plt.legend()
```
